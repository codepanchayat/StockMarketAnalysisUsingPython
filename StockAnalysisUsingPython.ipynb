```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
import plotly.graph_objects as go
import plotly.express as px
```

```python
today = date.today()
```

```python
d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today()-timedelta(days=365)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
```

```python
data = yf.download('INFY.NS', start=start_date, end=end_date, progress=False)
```

```python
data["Date"] = data.index
```

```python
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
print(data.head())
```

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"], open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"] )])
```

```python
figure.update_layout(title = "Infosys Stock Price Analysis", xaxis_rangeslider_visible=True)
figure.show()
```

```python
figure = px.bar(data, x = "Date", y= "Close")
figure.show()
```

```python
figure = px.line(data, x='Date', y='Close')
figure.update_xaxes(rangeslider_visible=True)
figure.show()
```

```python
figure = px.line(data, x='Date', y='Close', 
                 title='Stock Market Analysis with Time Period Selectors')

figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure.show()
```
